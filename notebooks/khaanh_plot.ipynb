```python
import matplotlib.pyplot as plt

# Ngày và độ chính xác
dates = ['May 22', 'May 29', 'June 5', 'June 12', 'June 19']
accuracies = [5, 0, 0, 0, 0]  # phần trăm độ chính xác

# Vẽ biểu đồ cột
plt.figure(figsize=(6, 4))
plt.bar(dates, accuracies)
plt.ylim(0, 100)
plt.ylabel('Accuracy (%)')
plt.title('Weekly Performance Accuracy of the Foundation Agents Algorithm')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
